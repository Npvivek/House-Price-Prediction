```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

## Data Exploration

```python
traindf = pd.read_csv('train.csv')
testdf = pd.read_csv('test.csv')
```

```python
sb.histplot(traindf['SalePrice'], kde = True)
plt.title('Distribution Of SalesPrice')
plt.show()
```

```python
# Filter out the numerical columns
numeric_df = traindf.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(50, 30))
sb.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Filter out only the numeric columns
numeric_df = traindf.select_dtypes(include=[np.number])

# Optionally, fill missing values in numeric columns (if any)
numeric_df = numeric_df.fillna(0)  # You can also choose other methods to handle NaNs

# Calculate the correlation matrix for numeric columns
corr_matrix = numeric_df.corr()

# Extract the correlation of SalePrice with all other features
saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)

# Display features with positive correlation to SalePrice
positive_corr_features = saleprice_corr[saleprice_corr > 0]
print("Features positively correlated with SalePrice:")
print(positive_corr_features)

# Display features with negative correlation to SalePrice
negative_corr_features = saleprice_corr[saleprice_corr < 0]
print("Features negatively correlated with SalePrice:")
print(negative_corr_features)
```
