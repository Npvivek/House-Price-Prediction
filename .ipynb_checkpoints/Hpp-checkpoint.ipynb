```python
# Function to identify low variance categorical features
def identify_low_variance_features(df, threshold=0.95):
    """
    Identify categorical features where the most common value dominates the feature
    based on a specified threshold.
    
    Parameters:
    df (DataFrame): The dataframe containing the data
    threshold (float): The threshold above which the feature will be considered low variance
    
    Returns:
    List: List of features that are considered low variance
    """
    low_variance_features = []
    
    for col in df.select_dtypes(include=['object']).columns:
        most_common_percentage = df[col].value_counts(normalize=True).max()
        if most_common_percentage > threshold:
            low_variance_features.append(col)
            print(f"Feature '{col}' is dominated by the value '{df[col].mode()[0]}' "
                  f"which appears in {most_common_percentage * 100:.2f}% of the data.")
    
    return low_variance_features

# Apply the function to the training data
low_variance_features = identify_low_variance_features(train)

print("\nLow variance features identified:", low_variance_features)
```

# Data Cleaning and Preprocessing

## Dropping columns 
* These columns are dropped because they likely contain too many missing values or might not be significantly relevant to the prediction of house prices. For instance:
  * Alley, PoolQC, MiscFeature: These features might have a high percentage of missing values, making them less reliable.
  * Utilities, Street: These might have little variance (e.g., most houses might have the same type of utilities or street access), offering little value to the model.

```python
columns=['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature','Utilities','Street']
train=train.drop(columns,axis=1)
test=test.drop(columns,axis=1)
```

## Filling missing values
* Categorical Variables: For categorical features (a and c), the missing values are filled with the most frequent category (mode). This is because the mode represents the most common case and is a reasonable assumption when data is missing.
* Numerical Variables: For numerical features (b and d), missing values are filled with the mean value. The mean is chosen to avoid skewing the data too much while maintaining consistency across records.

```python
a=['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','GarageType','GarageFinish','GarageQual','GarageCond']
for i in a:
    train[i]=train[i].fillna(train[i].mode()[0])
    test[i]=test[i].fillna(test[i].mode()[0])

c=['MSZoning','Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType']
for i in c:
    test[i]=test[i].fillna(test[i].mode()[0])
```

```python
b=['LotFrontage','MasVnrArea','GarageYrBlt']
for i in b:
    train[i]=train[i].fillna(train[i].mean())
    test[i]=test[i].fillna(test[i].mean())

d=['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageCars','GarageArea']
for i in d:
    test[i]=test[i].fillna(test[i].mean())
```

## Feature Engineering 

* TotalSF: Total square footage is a strong predictor of house price, as larger homes generally cost more. This feature sums up the square footage of different parts of the house.
* Age: The age of the house at the time of sale is relevant because newer houses might sell for more due to fewer required repairs or updates.
* RemodelAge: Indicates whether the house has been remodeled since it was built. This could positively impact the price, as remodeled homes are often more desirable.

```python
train['TotalSF']=train['TotalBsmtSF']+train['1stFlrSF']+train['2ndFlrSF']
test['TotalSF']=test['TotalBsmtSF']+test['1stFlrSF']+test['2ndFlrSF']
train['Age']=train['YrSold']-train['YearBuilt']
test['Age']=test['YrSold']-test['YearBuilt']
train['RemodelAge']=train['YearRemodAdd']!=train['YearBuilt'].astype(int)
test['RemodelAge']=test['YearRemodAdd']!=test['YearBuilt'].astype(int)
```

## Label Encoding
* Categorical variables are converted into numeric values using LabelEncoder. This is necessary because most machine learning models cannot directly handle categorical data and require numeric input.
  * Example: If a column MSZoning has values RL, RM, FV, they would be converted to 0, 1, 2, respectively.

```python
le=LabelEncoder()
for i in train.columns:
    if train[i].dtype=='object':
        train[i]=le.fit_transform(train[i])
for i in test.columns:
    if test[i].dtype=='object':
        test[i]=le.fit_transform(test[i])
```

## Target Variable Transformation

The SalePrice is log-transformed to reduce skewness. Housing prices often have a right-skewed distribution, and log transformation makes the data more normally distributed, which can improve the performance of linear models.

```python
train['SalePrice']=np.log1p(train['SalePrice'])
```

## Train-Test Split

The dataset is split into training and validation sets using a 70-30 ratio.

```python
x=train.drop(['SalePrice'],axis=1)
y=train['SalePrice']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state = 42)
```

## Model Training and Evaluation

* Random Forest: Train a Random Forest model, an ensemble of decision trees, on the training data. Random Forests are robust against overfitting and can capture complex relationships between features and the target variable.

* Linear Regression: Purpose: Train a simple Linear Regression model for comparison. This model assumes a linear relationship between the features and SalePrice. While it may not perform as well as Random Forest on complex datasets, it provides a good baseline.
 
* Evaluation: RMSE and R-squared metrics are calculated to assess model performance. RMSE indicates the average prediction error, while R-squared shows how much variance in SalePrice is explained by the model.

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
acc=rf.score(x_test,y_test)
print(acc)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
acc=lr.score(x_test,y_test)
print(acc)
```

## Predictions and Summary

The Random Forest model, performed better, is used to make predictions on the test set. The predictions are then exponentiated to revert the earlier log transformation, ensuring the predictions are on the original scale of SalePrice.

```python
submission=pd.DataFrame()
submission['Id']=test['Id']
final_predictions=rf.predict(test)
final_predictions=np.exp(final_predictions)
submission['SalePrice']=final_predictions
submission.to_csv('submission.csv',index=False)
```

### Why Use Random Forest and Linear Regression?
Random Forest:

* Strengths:
  * Handles high-dimensional data well.
  * Can capture nonlinear relationships between features and SalePrice.
  * Less prone to overfitting compared to a single decision tree due to averaging.
* Applicability: Ideal for this competition where the relationship between house features and prices is likely complex and nonlinear.

  
Linear Regression:

* Strengths:
  * Simple and interpretable.
  * Provides a good baseline for comparison.
* Applicability: Useful for understanding the linear relationships in the dataset and comparing against more complex models like Random Forest.
